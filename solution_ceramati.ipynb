```python
from concurrent.futures import ThreadPoolExecutor

def scrape_job(job):
    transformed_job = {
        'Department_name': job['department'],
        'title': job['name'],
        'location': job['location']['city'],
        'Job_ID'  : job[ 'jobAdId'], 
        'Experience': job['experienceLevel'],
        'description': [],
        'qualification': [],
        'posted by': job['company']['name']
    }
    return transformed_job

transformed_data = {}

with ThreadPoolExecutor() as executor:
    for departments, departmen_data in smart_recruiter_result.items():
        transformed_data[departments] = []
        results = executor.map(scrape_job, departmen_data['content'])
        for result in results:
            transformed_data[departments].append(result)

print(transformed_data)
```

```python
for department, jobs in transformed_data.items():
    print(f"{department}: [")
    for job in jobs:
        print("  {")
        for key, value in job.items():
            print(f"    {key}: {value}")
        print("  },")
    print("],")
```

```python
import json

with open('solution.json', 'w') as f:
    json.dump(transformed_data, f, indent=4)
```
